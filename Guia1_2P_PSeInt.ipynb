>> Las estructuras anidadas tienen las siguientes caracteristicas:
>>
>> - La primera condicion, se denomina _principal_ o _condicion raíz_
>> - Las demas condiciones se escribiran en la parte _falsa_ o _verdadera_ de la condicion que la antecede.
>> - La ultima condicion no es necesaria escribirla explicitamente.
>> - Si lo que se evalua es un intervalo con varios rangos, cada uno de estos debe ir en un lado de la condicion y sera lo mismo para las restantes.

# Proceso para desarrollar un algoritmo
Todo programa, independiente de su complejidad requiere para su solucion identificar, tres componentes:

- Entradas o inputs
- Proceso
- Salida o outputs

> ## Entradas o Inputs
>> Son las variables o constantes necesarias para que el programa funcione, responde a la pregunta: __Que datos necesita el progama__?
> ## Proceso
>>  Es la operacion o acciones que ejecutara el programa una vez haya recbido las entradas.
> ## Salida o Outputs
>> Es el resultado final del programa, este puede representarse mediante un mensaje, un grafico, un sonido, un sensor, etc.

# Ejemplo desarrollado en clase
Plantee un algoritmo que solicite dos numeros e indique cual de estos es mayor.
> ## Entradas
>> - primerNumero
>> - segundoNumero
> ## Proceso
>> Establecer condicion: primerNumero > segundoNumero
> ## Salidas
>> - True: "El primer numero es mayor"
>> - False: "El segundo numero es mayor"

# Desarrollo Ejemplo
En PSeInt se requieren las siguientes instrucciones:
>
    Algoritmo Ejemplo004
          Escribir "Dime un numero"
          Leer primerNumero
          Escribir "Dime otro numero"
          Leer segundoNumero
          Si primerNumero > segundoNumero Entonces
            Escribir "El primero es mayor"
          Sino
            Escribir "El segundo es mayor"
          FinSi
    FinAlgoritmo
>
> si lo piensa con un poco de detenimiento -o lo prueba-, se dará cuenta de que ese planteamiento no es correcto del todo: si introduce dos números iguales, le dirá que el segundo es el mayor; lo solucionaremos un poco más adelante.
>
> Se debe recordar que en el planteamiento de las condiciones se debe hacer un correcto uso de los operadores relacionales.

# Corrección 
Un buen programador debe considerar _todos los casos de uso_, o de otra forma, las distintas posibilidades que el usuario final puede elegir, el planteamiento anterior no incluia la situacion donde se escribieran dos numeros iguales, lo cual es perfectamente posible.

Para solucionarlo una alternativa es reescribir el codigo de la siguiente manera:

> 
    Algoritmo Ejemplo004c
        Escribir "Dime un numero"
        Leer primerNumero
        Escribir "Dime otro numero"
        Leer segundoNumero
        Si primerNumero > segundoNumero Entonces
            Escribir "El primero es mayor"
        Sino
            Si primerNumero < segundoNumero Entonces
                Escribir "El segundo es mayor"
            Sino
                Escribir "Son iguales"
            FinSi
        FinSi
    FinAlgoritmo
