# Algoritmos guia1

La siguiente actividad interactiva te permitirá familiarizarte con el concepto de algoritmos, eje central de la programación, además te presentará algunos concepto de manera implicita aun desconocidos para ti, con el pasar de las clases podras asociarlos a un vocabulario mas tecnico.

## Actividad 1 - Jugando con bloques

Emplearemos __Code.org__, una plataforma de acceso libre, donde no requieres cuentas para comenzar a programar con bloques, através de diferentes retos donde el nivel de dificultad aumenta a medida que avanzas.

1. Ingresa en el siguiente enlace:

> [Flappy Bird - Reto 1](https://studio.code.org/flappy/1)

2. La idea es mover el objeto (sprite), en este caso un pájaro, siguiendo las instrucciones y agregando los bloques correctos.
3. La pagina esta en ingles, por favor no la traduzcas, este idioma es la base de la programacion, por ahora aca tienes las Instrucciones en español:

> 1. Coloca un bloque en el bloque "al hacer clic" y pulsa "Ejecutar". Haz clic o toca la pantalla para mover a Flappy al objetivo.
> 2. Ahora hemos añadido algo de terreno. Añade un bloque al bloque "al tocar el suelo". Tu código debe ejecutar "FINALIZAR JUEGO" cuando Flappy se bloquea.
> 3. El bloque "Al ejecutar" te permite ejecutar código al iniciar el juego. Intenta configurar la velocidad del nivel y aletear hacia el objetivo.
> 4. Flappy volará a través de los obstáculos a menos que escribas código para que se estrelle. Agrega un bloque al bloque "al chocar con un obstáculo". Tu código debe ejecutar "FIN DEL JUEGO" cuando Flappy se estrelle contra el primer conjunto de tuberías.
> 5. Agreguemos un marcador y contemos puntos cuando Flappy supere obstáculos. Añadamos el bloque "al pasar un obstáculo" y luego anotemos un punto volando con Flappy por el primer conjunto de tuberías.
> 6. Vea si aún puede sumar un punto después de hacer que Flappy aletee una cantidad menor o mayor.
> 7. A continuación, intenta cambiar la escena colocando otro bloque en el bloque "cuando se ejecuta".
> 8. Puedes configurar efectos visuales en cualquier momento, no solo al inicio del juego. Intenta configurar una escena ALEATORIA cuando ocurran otros eventos, por ejemplo, al pasar un obstáculo.
> 9. Para variar las cosas, cuando choques contra un obstáculo, en lugar de finalizar el juego, intenta restablecer el puntaje a 0.
> 10. Crea tu propio juego de Flappy. Puedes cambiar todos los gráficos y las reglas, incluso la gravedad. Cuando termines, haz clic en "Finalizar" para que tus amigos lo prueben en sus teléfonos.

4. Termina el reto una vez lo completes llama al profesor para obtener tu primer sello, debes escribir lo siguiente en tu cuaderno:

> - Nombre de la actividad: Algoritmos reto en code.org - flappy
> - Ámbito conceptual: Los algoritmos en la vida común y en la programación
> - Competencia: 1
> - Objetivo: Emplear la plataforma code.org para cumplir un reto de programación que sirva como inicio a conceptos asociados a los algoritmos

5. Haz tu propia version del juego de Flappy, una vez lo tengas, contesta las siguientes preguntas en tu cuaderno, no es necesario que escribas el enunciado:
- Que pasa si intercambias el orden de dos bloques. Por ejemplo si en el primer paso en lugar de dejar `flap` ubicas el bloque `play sound`?
- Que hace que el juego arranque?
- En que momento empieza a moverse el personaje?
- Cual bloque hace que se cambie el escenario donde vive floppy?
- Que pasa si no colocas nada en el bloque `when hit an obstacle`
- Que hace el boton `run`?
